## Constructing a Portfolio

### Select a Candidates for the Portfolio

```python
djia = ['AXP','BA','CAT','CSCO','CVX','DD','DIS','GE', \
        'GS','HD','IBM','INTC','JNJ','JPM','KO','MCD', \
        'MMM','MRK','MSFT','NKE','PFE','PG','T','TRV', \
        'UNH','UTX','V','VZ','WMT','XOM']

small_portfolio = ['GOOG','F','AAPL','XOM','PG','KO']

portfolio = djia
```

### Download Historical Data

```python
import pandas as pd
import datetime
import pandas.io.data

end = datetime.datetime.today().date()
start = end - datetime.timedelta(3*365)

adjclose = pd.DataFrame()
for s in portfolio:
    adjclose[s] = pd.io.data.DataReader(s,"yahoo",start,end)['Adj Close']
```

### Equally Weighted Portfolio

```python
P_equal = pd.Series(0,index = adjclose.index)
for s in portfolio:
    P_equal += 100.0*adjclose[s]/adjclose[s][0]/len(portfolio)

figure(figsize=(12,6))

for s in portfolio:
    (100.0*adjclose[s]/adjclose[s][0]).plot(lw=0.3)

P_equal.plot(color='r',lw=4)
ylabel('Value');
title('Value of an equally weighted portfolio')
```

### Compute Component Returns

```python
returns = 252.0*log(adjclose).diff()

rmean = returns.mean()
mad = abs(returns-rmean).mean()

for s in returns.columns:
    for t in returns.index:
        if isnan(returns[s][t]):
            returns[s][t] = rmean[s]

figure(figsize = (12,0.3*len(rmean)))
subplot(1,2,1)
rmean.plot(kind='barh')
title('Mean Annualized Returns');

subplot(1,2,2)
mad.plot(kind='barh')
title('Mean Absolute Difference');
```

```python
import pulp

# The dataframes are indexed by timestamp. Unfortunately, this results
# in colons appearing the pulp variable names which are not processed
# properly by GLPK and possibly other solvers. So we create a dictionary
# with values of epoch seconds used for pulp variable names.

tsec = {t:"{:%s}".format(t) for t in returns.index}

# Decision variables
#   m: mean absolute deviation 
#   u: fraction of the portfolio in risk-free asset
#   w[s]: portfolio weights
#   y[t],z[t]: absolute deviation at time t

m = pulp.LpVariable('m',lowBound = 0)
u = pulp.LpVariable('u',lowBound = 0)
w = pulp.LpVariable.dicts('w',portfolio,lowBound = 0)
y = pulp.LpVariable.dicts('y',tsec.values(),lowBound = 0)
z = pulp.LpVariable.dicts('z',tsec.values(),lowBound = 0)
```

## Problem 1: Solve for Dominating MAD Portfolio

```python
lp = pulp.LpProblem('MAD Portfolio',pulp.LpMinimize)
lp += m    
lp += m == pulp.lpSum([(y[t] + z[t])/float(len(returns.index)) for t in tmap.values()])
lp += pulp.lpSum([w[s] for s in symbols]) == 1.0

for t in returns.index:
    lp += y[tsec[t]] - z[tsec[t]] == pulp.lpSum([w[s]*(returns[s][t]-rmean[s]) for s in symbols]) 
        
lp.solve()
m_min = m.varValue
m_min
```

```python
# Solve for maximum return at minimum MAD

r = pulp.LpVariable('r',lowBound = 0)

lp = pulp.LpProblem('MAD Portfolio',pulp.LpMaximize)
lp += r
lp += r == pulp.lpSum([w[s]*rmean[s] for s in symbols])
lp += m_min == pulp.lpSum([(y[t] + z[t])/float(len(returns.index)) for t in tmap.values()])
lp += pulp.lpSum([w[s] for s in symbols]) == 1.0
for t in returns.index:
    lp += y[tsec[t]] - z[tsec[t]] == pulp.lpSum([w[s]*(returns[s][t]-rmean[s]) for s in symbols]) 
        
lp.solve()
r_min = r.varValue
w_min = pd.Series([pulp.value(w[s]) for s in symbols], index= symbols)
```

```python
figure(figsize = (15,0.3*len(rmean)))
subplot(1,2,1)
w_min.plot(kind='barh')
subplot(1,2,2)
w_nz = w_min[w_min > 0]
pie(w_nz,labels=w_nz.index);
axis('equal');
```
