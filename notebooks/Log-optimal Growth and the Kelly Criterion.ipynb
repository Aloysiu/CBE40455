### Opportunity to Manage for Optimal Growth

Consider a continuing 'investment opportunity' for which, at each stage, an invested dollar will yield either two dollars with 
probability $p$ or nothing with probability $1-p$.

<img src="files/images/Kelly_Criterion_Fig1.png" style = "width:150px">

Whatever investment strategy we choose, at each stage $k$ there will be a total return $R_k$. Starting with a wealth $W_0$, after $k$ stages our wealth will be

$$W_k = R_kR_{k-1}\cdots R_2R_1W_0$$

Now let's consider a specific investment strategy in which we invest a fraction $\alpha$ of our total wealth into the risky opportunity, and retain a fraction $1-\alpha$ for future use. The return $R_k$ is given by

$$R_k = \begin{cases} 1+\alpha & \mbox{with probability}\quad p \\ 1-\alpha & \mbox{with probability}\quad 1-p\end{cases}$$

What should be our strategy for choosing $\alpha$?

### Maximizing Expected Wealth

At first glance, maximizing expected wealth would seem like a good investment objective. Suppose  after $k$ stages we have witnessed $u$ profitable outcomes (i.e., 'wins'), and $k-u$ outcomes showing a loss. The remaining wealth will be given by

$W_k/W_0 = (1+\alpha)^u(1-\alpha)^{k-u}$

where the binomial distribution gives the probability of observing $u$ 'wins' in $k$ trials

$${k\choose{u}}p^u (1-p)^{k-u}$$

The expected value of $W_k$ is given by

$$E[W_k/W_0] = \sum_{u=0}^{k} {k\choose{u}}p^u (1-p)^{k-u}(1+\alpha)^u(1-\alpha)^{k-u}$$

Next we plot $E[W_k/W_0]$ as a function of $\alpha$.

```python
from scipy.misc import comb

K = 50
p = 0.6

for alpha in linspace(0,1,300):
    w = sum([comb(K,u)*((p*(1+alpha))**u)*(((1-p)*(1-alpha))**(K-u)) for u in range(0,K+1)])
    plot(alpha,w,'b.')

xlabel('alpha')
ylabel('E[W(k)/W(0)]')
title('Expected fraction of remaining wealth after {:d} trials'.format(K))
```

The following cell simulates the behavior of this process for given parameter values. Try different values of $\alpha$ in the range from 0 to 1 to see what happens.

```python
K = 50
p = 0.6
alpha = 0.8

# Compute an array of future returns
R = array([1-alpha + 2*alpha*float(random.uniform() <= p) for _ in range(0,K)])

# Use returns to compute fraction of wealth that remains
W = concatenate(([1.0],cumprod(R)))

# plot
plot(W)
xlabel('Stage k')
ylabel('Fraction of Initial Wealth');
```

Attempting to maximize wealth leads to a risky strategy where all wealth is put at risk at each stage hoping for a string of $k$ wins.  The very high rewards for this one outcome mask the fact that the most common outcome is to lose everything. If you're not convinced of this, go back and run the simulation a few more times for values of alpha in the range 0.8 to 1.0.  

If $\alpha =1$, the probability of still having money after $k$ stages is $(1-p)^k$.

```python
semilogy(range(0,K+1), [(1-p)**k for k in range(0,K+1)])
title('Probability of non-zero wealth after k stages')
xlabel('Stage k')
ylabel('Probability')
grid();
```

### Maximizing Growth

To acheive a different result we need to consider optimization objective that incorporates a measure of risk. For example, the log ratio of current to starting wealth gives a relationship

$$\ln W_k/W_0 = \sum_{k'=1}^k \ln R_{k'}$$

Subjectively, the log ratio focuses relative rather than absolute growth which, for many investors, is a better indicator of investment objectives. Rather than risk all for an enormous but unlikely outcome, a strategy optimizing expected  relative growth will tradeoff risky strategies for more robust strategies demonstrating relative growth.  

Taking expectations

$$E[\ln W_k/W_0] = \sum_{k'=1}^k E[\ln R_{k'}] = k E[\ln R_{k'}]$$

where

$$E[\ln R_{k'}] = p\ln(1+\alpha) + (1-p)\ln(1-\alpha)$$

With simple calculus we can show that maximizing $E[\ln W_k/W_0]$ requires

$$\alpha = 2p-1$$

which yields a growth rate per stage

$$m = E[\ln R_{k'}] = \ln 2 + p\ln(p) + (1-p)\ln(1-p)$$



## Kelly's Criterion

<img src="files/images/Kelly_Criterion_Fig2.png" style = "width:150px">

$$E[\ln R_{k'}] = p\ln(1+ b\alpha) + (1-p)\ln(1-\alpha)$$

Solving for $\alpha$

$$\alpha = \frac{p(b+1)-1}{b}$$

## Volatility Pumping

<img src="files/images/Kelly_Criterion_Volatility_Pumping.png" style = "height:120px">


```python
# payoffs for two states
u = 1.1
d = 1/u
p = 0.49

K = 100

Z = array([float(random.uniform() <= p) for _ in range(0,K)])
R = d + (u-d)*Z
S = cumprod(concatenate(([1],R)))

ElnR = p*log(u) + (1-p)*log(d)
print "Expected return = {:0.5}".format(ElnR)

#semilogy(S);

ElnR = lambda alpha: p*log(alpha*u +(1-alpha)) + (1-p)*log(alpha*d + (1-alpha))

a = linspace(0,1)

#plot(a,map(ElnR,a))

from scipy.optimize import fminbound
alpha = fminbound(lambda(alpha): -ElnR(alpha),0,1)
print alpha

#plot(alpha, ElnR(alpha),'r.',ms=10)

R = alpha*d + (1-alpha) + alpha*(u-d)*Z
S2 = cumprod(concatenate(([1],R)))
plot(range(0,K+1),S,range(0,K+1),S2)
legend(['Stock','Stock+Cash'])
```
