## Computing Returns

Defining return as

$$R_{k} = \frac{S_k - S_{k-1}}{S_{k-1}}$$

Given the price data stored as a pandas `DataSeries` `S`, a corresponding data series for returns is computed using the pandas `shift()` function, followed by the `dropna()` function to remove missing values.

```python
R = ((S - S.shift(+1))/S.shift(+1)).dropna()

figure(figsize=(10,6))
subplot(2,1,1)
S.plot()
ylabel('Price')
title(description)

subplot(2,1,2)
R.plot()
ylabel('Return')
```

```python
df = pd.DataFrame({'S':S,'R':R}, index = R.index)
```

## Statistical Properties of the Returns

A basic assumption in developing developing stochastic price models is that the residuals are indepdendent and identically distributed (i.i.d.) random variates.  Here we show the results of several common statistical tests that would screen out non-i.i.d. random variates.

### Lag Plot of $R_{k+1}$ versus $R_k$

```python
plot(R[0:-1],R[1:],'.')
axis('equal');
xlabel('R(k)')
ylabel('R(k+1)')
```

### Partial Autocorrelation

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(R,lags=min(30,len(R)))
xlabel('Lag');
```

### Distribution of First Half versus Second Half of the Data Set

```python
from scipy.stats import t

nbins = min(100,int(sqrt(len(R))))
k = int(len(R)/2)

r = linspace(R.min(),R.max())

figure();

param = t.fit(R[:k])
R[:k].hist(bins=nbins,normed=1,alpha=0.3,color='r')
plot(r,t.pdf(r,df=param[0],loc=param[1],scale=param[2]),'r-',lw=3);

R[k:].hist(bins=nbins,normed=1,alpha=0.3,color='g')
param = t.fit(R[k:])
plot(r,t.pdf(r,df=param[0],loc=param[1],scale=param[2]),'b-',lw=3);

legend(['R[:k]','R[k:]'])
```

## Bootstrap Forecasting

We have computed return as

$$R_{k} = \frac{S_k - S_{k-1}}{S_{k-1}}$$

and executed several tests of the hypothesis that $R_k$ behaves as independent, identically distributed (i.i.d.) random variable. If we're satisfied with this hypothesis, then we can randomly sample the historical return data $R_k$ to simulate future prices.  This bootstrap technique is relatively simple to implement as shown below.

```python
from scipy.stats import lognorm

N = 1000
T = 252
h = 1

figure(figsize=(10,6))
subplot(2,1,1)

plot(S)
title(description)
xlabel('Trading Days')

# Initialize P using the last available price
P = array([S[-1]]*N)

# generator of random variates
Zgenerator = (array([random.choice(R) for _ in xrange(N)]) for k in range(0,T))

k = len(S)
for Z in Zgenerator:
    k += 1
    P += P*Z
    plot([k]*N,P,'b.',ms=0.1)

subplot(2,1,2)
nbins = min(100,int(1.5*sqrt(N)))
hist(P,bins=nbins,normed=1,alpha=0.3,color='b');

shape,loc,scale = lognorm.fit(P,floc=0)
print shape,loc,scale
x=np.linspace(0,max(P),100)
pdf_fitted = lognorm.pdf(x, shape, loc=loc, scale=scale) # fitted distribution
plot(x,pdf_fitted,'b-',lw=3)

xlabel('Final Price')
ylabel('Probability');
title(description)
```

## Geometric Brownian Motion

$$S_{k+1} = S_k + \alpha h S_k + \beta \sqrt{h} S_k Z_k$$

where $Z_k \sim N(0,1)$ and $h$ corresponds to a sampling period of 1 'trading day'. There are typically 252 trading days in a calendar year, 63 trading days in a quarter, and 21 trading days in a month.

Defining return as

$$R_{k} = \frac{S_k - S_{k-1}}{S_{k-1}} = \alpha h + \beta \sqrt{h} Z_k$$

then the statistical model for $R_k$ becomes

$$R_{k} = \alpha h + \beta \sqrt{h} Z_k$$

For the case of Geometric Brownian Motion the hypothesis is that $R_k$ is a random variable drawn from a the normal distribution $R_k = \sim N(\alpha h, \beta\sqrt{h}$).

```python
from scipy.stats import norm

figure(figsize=(10,4))
plot(S)
title(description)
xlabel('Trading Days')

N = 200    # number of realizations
T = 252    # simulation time horizon in trading days
h = 1      # time step in trading days

param = norm.fit(R)
alpha = param[0]/h
beta = param[1]/sqrt(h)

# Initialize P
P = array([S[-1]]*N)

# generator of random variates
Zgenerator = (array(norm.rvs(loc=alpha*h,scale=beta*sqrt(h),size=N)) for k in range(0,T))

k = len(S)
for Z in Zgenerator:
    k += 1
    P += P*Z
    plot([k]*N,P,'b.',ms=0.1)
```

### Forecast Assuming a Normal Distribution of Returns

```python
from scipy.stats import norm,lognorm

param = norm.fit(R)
alpha = param[0]
beta = param[1]

N = 10000
T = 252
h = 1

# Initialize P
Pgbm = array([S[-1]]*N)

# generator of random variates
Zgenerator = (array(norm.rvs(loc=alpha*h,scale=beta*sqrt(h),size=N)) for k in range(0,T))

for Z in Zgenerator:
    Pgbm += Pgbm*Z
    
figure(figsize=(10,4))
nbins = min(100,int(1.5*sqrt(N)))
hist(Pgbm,bins=nbins,normed=1,alpha=0.4,color='b');

shape,loc,scale = lognorm.fit(Pgbm,floc=0)
print shape,loc,scale
x=np.linspace(0,max(Pgbm),100)
pdf_fitted = lognorm.pdf(x, shape, loc=loc, scale=scale) # fitted distribution
plot(x,pdf_fitted,'b-',lw=3)
xlabel('Final Price')
ylabel('Probability');
title(description)
```

### Forecast Assuming a t-Distribution of Returns

```python
from scipy.stats import t,lognorm

param = t.fit(R)
df = param[0]
alpha = param[1]
beta = param[2]

N = 10000
T = 252
h = 1

# Initialize P
Pgtm = array([S[-1]]*N)

# generator of random variates
Zgenerator = (array(t.rvs(df,loc=alpha*h,scale=beta*sqrt(h),size=N)) for k in range(0,T))

for Z in Zgenerator:
    Pgtm += Pgtm*Z

figure(figsize=(10,4))
nbins = min(100,int(1.5*sqrt(N)))
hist(Pgtm,bins=nbins,normed=1,alpha=0.4,color='r');

shape,loc,scale = lognorm.fit(Pgtm,floc=0)
print shape,loc,scale
x=np.linspace(0,max(Pgtm),100)
pdf_fitted = lognorm.pdf(x, shape, loc=loc, scale=scale) # fitted distribution
plot(x,pdf_fitted,'r-',lw=3)
xlabel('Final Price')
ylabel('Probability');
title(description)
```
